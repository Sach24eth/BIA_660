```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import wordcloud 
from wordcloud import WordCloud, STOPWORDS
import re
import string
import math
import seaborn as sns
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import tokenize
```

```python
data = pd.read_csv("HotelReview_Clean_v1.1.csv")
type(data.review_text)
data.head
```

```python
#This is paragraph wise ,
# a few thiings here, the compund score for the paragraphs is kinda giving us good results
# comparing to the user given rating we can see that there are clear cut values for compund score.
# Issue here is that: Tricky sentences- > show example of what tricky means
reviews= data.review_text;
sid = SentimentIntensityAnalyzer()

for review in reviews:
    line_list=tokenize.sent_tokenize(review)
    print("\nReview:\n",review)
    sid = SentimentIntensityAnalyzer()
    print("\nSIA scores:\n",sid.polarity_scores(review))
```

```python
# This is per sentence wise, rating for each review we need to look at how we can use the compound score and combine it in a way
#scrap
for review in data.review_text:
    line_list=tokenize.sent_tokenize(review)
    print(review)
    negetive=0;
    positive=0;
    compound=0;
    neutral=0;
    sid = SentimentIntensityAnalyzer()
    print("\nSentences: \n")
    for sentence in line_list:
        print(sentence)
        ss = sid.polarity_scores(sentence)
        negetive+=ss['neg']
        positive+=ss['pos']
        neutral+=ss['neu']
        compound+=ss['compound']
        
        for k in sorted(ss):
            print('{0}: {1}, '.format(k, ss[k]))
        print("\n")
#         print("Negetive:", negetive)
    print("Negetive sum for each review: ", negetive)
    print ("Positive sum for each review: ", positive)
    print ("Neutral sum for each review: ", neutral)
    print ("Compound score sum for each review ", compound)
    print(sid.polarity_scores(review))
```
